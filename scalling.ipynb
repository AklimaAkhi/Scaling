# Standardization

```python
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
stds= StandardScaler()
```

```python
df1['Marketing Spend']= stds.fit_transform(df1[['Marketing Spend']])
df1['Administration']= stds.fit_transform(df1[['Administration']])
df1['Transport']= stds.fit_transform(df1[['Transport']])
```

# RobustScaler
